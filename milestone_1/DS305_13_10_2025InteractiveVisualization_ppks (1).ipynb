#Happy Planet Index (HPI) Data Set:
We will use the Happy Planet Index (HPI) http://happyplanetindex.org/ dataset

ecological resources most efficiently to live long, happy lives. It is not only an interesting

```python
import altair as alt
```

```python
#Adding Zoom-In and Zoom-Out to a Static Scatter Plot
alt.Chart(hpi_df).mark_circle().encode(
x='Wellbeing (0-10):Q',
y='Happy Planet Index:Q',
color='Region:N',tooltip=['Country']).interactive()
```

```python
#Adding Zoom-In and Zoom-Out to a Static Scatter Plot
alt.Chart(hpi_df).mark_bar().encode(
x='Wellbeing (0-10):Q',
y='Happy Planet Index:Q',
color='Region:N',).interactive()
```

```python
# Adding Zoom-In and Zoom-Out to a Line Chart
alt.Chart(hpi_df).mark_line().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color='Region:N'
).interactive()
```

```python
# Adding Zoom-In and Zoom-Out to an Area Chart
alt.Chart(hpi_df).mark_area().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color='Region:N'
).interactive()
```

```python
# Adding Zoom-In and Zoom-Out to a Histogram
alt.Chart(hpi_df).mark_bar().encode(
    x=alt.X('Wellbeing (0-10):Q', bin=True),  # Binning for histogram
    y='count()',
    color='Region:N'
).interactive()
```

```python
#Adding Hover and Tooltip Functionality to a Scatter Plot
alt.Chart(hpi_df).mark_point().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color='Region:N',
    tooltip=['Country', 'Region', 'Wellbeing (0-10)', 'Happy Planet Index', 'Life Expectancy (years)']
).interactive()
```

```python
#Adding Hover and Tooltip Functionality to a Scatter Plot
alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color='Region:N',
    tooltip=['Country', 'Region', 'Wellbeing (0-10)', 'Happy Planet Index', 'Life Expectancy (years)']
).interactive()
```

```python
# Adding Hover and Tooltip Functionality to a Bar Plot
alt.Chart(hpi_df).mark_bar().encode(
    x='Region:N',  # Categorical data for the x-axis
    y='Happy Planet Index:Q',  # Quantitative data for the y-axis
    color='Region:N',  # Color each bar based on the region
    tooltip=['Country', 'Region', 'Wellbeing (0-10)', 'Happy Planet Index', 'Life Expectancy (years)']  # Tooltip with relevant information
).interactive()  # Allow interactivity such as zooming or panning
```

```python
alt.Chart(hpi_df).mark_area().encode(
    x='Wellbeing (0-10):Q',  # Quantitative data for the x-axis
    y='Happy Planet Index:Q',  # Quantitative data for the y-axis
    color='Region:N',  # Color by region
    tooltip=['Country', 'Region', 'Wellbeing (0-10)', 'Happy Planet Index', 'Life Expectancy (years)']  # Tooltip info
).interactive()

```

```python
alt.Chart(hpi_df).mark_bar().encode(
    x=alt.X('Wellbeing (0-10):Q', bin=True),  # Binned quantitative data for histogram
    y='count()',  # Count of records for each bin
    color='Region:N',  # Color by region
    tooltip=['count()', 'Wellbeing (0-10)', 'Region']  # Tooltip info
).interactive()
```

```python
#Exploring Select and Highlight Functionality on a Scatter Plot
selected_area = alt.selection_interval()
alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',color='Region:N').add_selection(selected_area)
```

```python
selected_area = alt.selection_interval()
alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color=alt.condition(selected_area, 'Region:N', alt.value('red'))).add_selection(selected_area)
```

```python
# multiple altair charts placed horizontally next to each other
chart = alt.Chart(hpi_df).mark_circle().encode(
y='Happy Planet Index',
color='Region:N')
chart1 = chart.encode(x='Wellbeing (0-10)')
chart2 = chart.encode(x='Life Expectancy (years)')
alt.hconcat(chart1, chart2)
```

```python
# hover and tooltip across multiple charts
selected_area = alt.selection_interval()
chart = alt.Chart(hpi_df).mark_circle().encode(
    y='Happy Planet Index',
    color=alt.condition(selected_area, 'Region', alt.
value('lightgray'))
).add_selection(
selected_area
)
chart1 = chart.encode(x='Wellbeing (0-10)')
chart2 = chart.encode(x='Life Expectancy (years)')
chart1 | chart2
```

```python
# Bar plots with hover and linked interaction
selected_area = alt.selection_interval()

# Base bar chart
chart = alt.Chart(hpi_df).mark_bar().encode(
    y='Happy Planet Index:Q',
    color=alt.condition(selected_area, 'Region:N', alt.value('lightgray')),
    tooltip=['Country', 'Region', 'Happy Planet Index', 'Wellbeing (0-10)', 'Life Expectancy (years)']
).add_selection(selected_area)

# First bar chart (Wellbeing vs Happy Planet Index)
chart1 = chart.encode(x='Wellbeing (0-10):Q')

# Second bar chart (Life Expectancy vs Happy Planet Index)
chart2 = chart.encode(x='Life Expectancy (years):Q')

# Combine both bar charts
chart1 | chart2
```

```python
input_dropdown = alt.binding_select(options=list(set(hpi_df['Region'])))
selected_points = alt.selection_single(fields=['Region'], bind=input_dropdown, name='Select')

color = alt.condition(
    selected_points,
    alt.Color('Region:N'),
    alt.value('lightgray')
)

alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color=color,
    tooltip=['Region:N']
).add_selection(
    selected_points
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample Data
df_sales = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Sales': [50000, 30000, 70000, 40000, 60000]
})

df_revenue = pd.DataFrame({
    'Month': pd.date_range('2023-01-01', periods=12, freq='M'),
    'Revenue': [12000, 15000, 17000, 14000, 20000, 22000, 21000, 25000, 23000, 24000, 22000, 26000]
})

df_market_share = pd.DataFrame({
    'Region': ['North', 'South', 'East', 'West'],
    'MarketShare': [35, 25, 20, 20]
})

df_customer_satisfaction = pd.DataFrame({
    'CustomerSatisfaction': [80, 60, 90, 75, 70],
    'Sales': [50000, 30000, 70000, 40000, 60000]
})

df_sales_distribution = pd.DataFrame({
    'SalesAmounts': [50000, 30000, 70000, 40000, 60000, 45000, 38000, 69000, 72000, 59000]
})
```

```python
# 1. Bar Chart: Sales by Product
plt.figure(figsize=(8, 5))
sns.barplot(x='Product', y='Sales', data=df_sales, palette='Blues_d')
plt.title('Sales by Product')
plt.ylabel('Sales ($)')
plt.xlabel('Product')
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Interactive Bar Chart: Sales by Product using Plotly
bar_chart = px.bar(df_sales, x='Product', y='Sales', title='Sales by Product',
                   labels={'Sales': 'Sales ($)', 'Product': 'Product'}, text_auto=True)

# Customize the appearance
bar_chart.update_traces(marker_color='lightblue', textposition='outside')
bar_chart.update_layout(title_font_size=20, xaxis_title_font_size=16, yaxis_title_font_size=16)

# Show the interactive chart
bar_chart.show()
```

```python
# 2. Line Chart: Monthly Revenue over Time
plt.figure(figsize=(8, 5))
plt.plot(df_revenue['Month'], df_revenue['Revenue'], marker='o', linestyle='-', color='b')
plt.title('Monthly Revenue Over Time')
plt.ylabel('Revenue ($)')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Interactive Line Chart: Monthly Revenue over Time using Plotly
line_chart = px.line(df_revenue, x='Month', y='Revenue', title='Monthly Revenue Over Time',
                     labels={'Revenue': 'Revenue ($)', 'Month': 'Month'})

# Customize the appearance
line_chart.update_traces(mode='lines+markers', marker=dict(size=8, color='blue'))
line_chart.update_layout(title_font_size=20, xaxis_title_font_size=16, yaxis_title_font_size=16)
line_chart.update_xaxes(tickangle=45)  # Rotate the x-axis labels

# Show the interactive chart
line_chart.show()
```

```python
# 3. Pie Chart: Market Share by Region
plt.figure(figsize=(6, 6))
plt.pie(df_market_share['MarketShare'], labels=df_market_share['Region'], autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Market Share by Region')
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Interactive Pie Chart: Market Share by Region using Plotly
pie_chart = px.pie(df_market_share, values='MarketShare', names='Region',
                   title='Market Share by Region',
                   color_discrete_sequence=px.colors.qualitative.Set2,  # Set the same color palette
                   hole=0.0)  # This can be adjusted for a donut chart effect if needed

# Add hover information (percentage automatically shown in Plotly)
pie_chart.update_traces(textinfo='percent+label', pull=[0.1, 0, 0, 0])  # Optional: 'pull' creates a separation effect

# Customize the layout
pie_chart.update_layout(title_font_size=20)

# Show the interactive chart
pie_chart.show()
```

```python
# 4. Scatter Plot: Customer Satisfaction vs Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(x='CustomerSatisfaction', y='Sales', data=df_customer_satisfaction, s=100, color='purple')
plt.title('Customer Satisfaction vs Sales')
plt.ylabel('Sales ($)')
plt.xlabel('Customer Satisfaction (%)')
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Interactive Scatter Plot: Customer Satisfaction vs Sales using Plotly
scatter_plot = px.scatter(df_customer_satisfaction, x='CustomerSatisfaction', y='Sales',
                          title='Customer Satisfaction vs Sales',
                          labels={'CustomerSatisfaction': 'Customer Satisfaction (%)', 'Sales': 'Sales ($)'},
                          size_max=10, color_discrete_sequence=['purple'])  # Set the marker color to purple

# Customize the marker size and style
scatter_plot.update_traces(marker=dict(size=10))

# Customize the layout
scatter_plot.update_layout(title_font_size=20, xaxis_title_font_size=16, yaxis_title_font_size=16)

# Show the interactive chart
scatter_plot.show()
```

## Sunburst Charts:
Sunburst charts represent hierarchial data in the form of concentric circles. Here the innermost circle is the root node which defines the parent, and then the outer rings move down the hierarchy from the centre. They are also called radial charts.We can use them to plot

* Worldwide mobile Sales where we can drill down as follows:   
    * innermost circle represents total sales  
    * first outer circle represents continentwise sales
    * second outer circle represents countrywise sales within each continent
    
    
* Disease outbreak hierarchy


* Real Estate Industrial chain

```python
##Example 4: Let us illustrate plot the

#Create a dictionary having a set of people represented by a character array and the parents of these characters represented in another
## array and the values are the values associated to the vectors.
data = dict(
    character=["Ramulu", "Rani", "Rajesh", "Ravali", "Ravi", "Laxmi", "Radha", "Kumar", "Babu"],
    parent=["", "Ramulu", "Ramulu", "Rajesh", "Rajesh", "Ramulu", "Ramulu", "Radha", "Ramulu" ],
    #value=[10, 14, 12, 10, 2, 6, 6, 4, 4]
    )

fig = px.sunburst(
    data,
    names='character',
    parents='parent',
    #values='value',
    title="Family chart"
)
fig.show()
```

```python
# Example 4: Plotting a sunburst chart for department hierarchy

# Create a dictionary representing departments and their parent courses with associated values.
data = dict(
    department=["IcfaiTech", "AI&DS", "AI", "CSE", "AI&ML", "ECE", "VLSI", "Mech", "Robotics", "Civil", "Bsc", "CS","Honors","Honors, Research", "DA","Honors","Honors, Research", "BCA","Honors","Honors, Research"],
    course=["", "IcfaiTech", "AI&DS", "IcfaiTech", "AI&DS", "IcfaiTech", "ECE", "IcfaiTech", "Mech", "IcfaiTech","CSE", "Bsc","CS","CS", "Bsc", "DA","DA","CSE","BCA","BCA"],
    #value=[10, 12, 12, 10, 2, 6, 6, 4, 4, 4, 4, 5, 10,10,10]
)

# Plot the sunburst chart
fig = px.sunburst(
    data,
    names='department',
    parents='course',  # Corrected 'courses' to 'parents' for hierarchical structure
    #values='value',
    title="IcfaiTech, Hyderabad"
)
fig.show()
```
